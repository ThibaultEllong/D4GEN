# Reading data

```python
RAW_DATA_PATH = Path("C:/Users/TUF15/Desktop/D4GEN/data/raw/raw_dataset_raw.csv")
```

```python
raw_dataset = pd.read_csv(RAW_DATA_PATH, delimiter=';')
```

```python
print ('Base data has %i rows and %i columns' % (raw_dataset.shape[0], raw_dataset.shape[1]))
```

# Cleaning data

```python
prepared_dataset = raw_dataset.copy()
```

```python
# Elimination des lignes vides
prepared_dataset = prepared_dataset.dropna(how='all')
prepared_dataset.shape
```

```python
# Elimination des lignes vides
prepared_dataset = prepared_dataset.dropna(how='all')

# Elimination des colonnes vides
prepared_dataset = prepared_dataset.dropna(how='all', axis=1)

# Normalisation des noms des colonnes
normalized_columns = []
for i in range(len(prepared_dataset.columns)):
    normalized_columns.append(str(prepared_dataset.columns[i]).replace(' ', '_').replace('é', 'e').replace('à', 'a').replace("'", '_').upper())
prepared_dataset.columns = normalized_columns

prepared_dataset.shape
```
